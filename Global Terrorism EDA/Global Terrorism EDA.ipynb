## Importing Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Importing Dataset 

```python
df=pd.read_csv('globalterrorismdb_0718dist.csv')
```

```python
df.head()
```

```python
df.shape
```

#### Displaying All the coulmns in the Dataset 

```python
pd.set_option('display.max_columns', 135)
df.head()
```

### Renaming All the Necessary Coulmns for Analysis 

```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region',
                     'provstate':'State','city':'City','latitude':'latitude','longitude':'longitude','summary':'summary',
                     'attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
                     'nwound':'Wound'},inplace=True)
```

```python
df.head()
```

### Seperating Necessary Coulmns used for Analysis 

```python
df_New=df[['Year','Month','day','Country','State','Region','City','latitude','longitude',"Attacktype","Targettype",'kill',
               'Wound','target1','summary','Group','Targettype','Weapon','motive']]
```

```python
df_New.head()
```

```python
df_New.info()
```

```python
df_New.describe()
```

```python
df_New.isnull().sum()
```

Though there are many null values but they cannot be handled before Data visualization cause if they handled  it will affect the Visualizations

### Correaltion between Numerical coulmns 

```python
df_New.corr()
```

```python
plt.figure(figsize=(18,10))
sns.heatmap(df_New.corr(),annot=True)
```

## Attacks Per year 

```python
Attacks_Per_Year = df_New['Year'].value_counts().sort_index()
Year = df_New['Year'].unique()
```

```python
plt.figure(figsize=(25,10))
sns.barplot(x = Year, y = Attacks_Per_Year, palette="deep")
plt.xlabel("Year",fontsize=25)
plt.ylabel("Attacks_Per_Year",fontsize=25)
```

We can clearly see that 2014 year had the most no of Terrorist Attacks 

## Regions Affected 

```python
plt.figure(figsize = (15,7))
df_New.groupby(['Region'])['kill'].sum().sort_values(ascending = False).plot(kind = 'bar', color = 'red')
plt.title('No. of people killed Region wise')
plt.ylabel("Killed")
```

We can say that Middle East & North Africa is the region which is most affected by Terrorism having almost 140000 casualties

## Top 15 Countries Affected 

```python
plt.figure(figsize=(25,10))
sns.barplot(df_New['Country'].value_counts()[:15].index,df_New['Country'].value_counts()[:15].values)
plt.title('Top 15 Countries Affected',fontsize=25)
plt.xlabel('Countries',fontsize=25)
plt.ylabel('No of Attacks',fontsize=25)
```

Iraq is the most affected country by Terrorism in the world

## Top 15 Cities Affected

```python
plt.figure(figsize=(25,10))
sns.barplot(df_New['City'].value_counts()[1:15].index,df_New['City'].value_counts()[1:15].values)
plt.title('Top 15 Cities Affected')
plt.xlabel('City',fontsize=25)
plt.ylabel('No of Attacks',fontsize=25)
```

Baghdad is the most affected city by Terrorism in the world

## Attack Type  

```python
plt.figure(figsize=(25,15))
df_New['Attacktype'].value_counts().plot.pie(autopct="%1.1f%%",colors=sns.color_palette('Set2'))
```

We can see that most of the attacks are Bombing/Explosion

## No. of people killed per Year 

```python
plt.figure(figsize = (25,10))
df_New.groupby(['Year'])['kill'].sum().plot(kind = 'bar',color='red')
plt.title('No. of people killed per Year',fontsize=25)
plt.xlabel("Year",fontsize=25)
plt.ylabel("Count of Killed people",fontsize=25)
```

Year 2014 had the most casulaties which was about 45000 people

##  No. of people Wounded per Year

```python
plt.figure(figsize = (25,10))
df_New.groupby(['Year'])['Wound'].sum().plot(kind = 'bar',color='orange')
plt.title('No. of people Wounded per Year',fontsize=25)
plt.xlabel("Year",fontsize=25)
plt.ylabel("Count of Wounded people",fontsize=25)
```

Year 2015 had most wounded people by Terrorism around the world

## No. of people killed Country wise

```python
plt.figure(figsize = (15,7))
df_New.groupby(['Country'])['kill'].sum().sort_values(ascending = False).head(15).plot(kind = 'bar',color='red')
plt.title('No. of people killed Country wise',fontsize=25)
plt.xlabel("Year",fontsize=25)
plt.ylabel("Killed",fontsize=25)
plt.show()
```

Due to Terrorism almost 80000 People had died in Iraq

##  Top 10 Terrorist Organisation with Highest No of Terrorist Attacks

```python
 len(np.unique(df_New['Group']))
```

As there are almost 3500 different organization we will see the top 10 

```python
plt.figure(figsize = (25,10))
sns.barplot(x = df_New['Group'].value_counts()[1:11].index ,y= df_New['Group'].value_counts()[1:11].values,palette='deep')
plt.title('Top 10 Terrorist Organisation with Highest No of Terrorist Attacks',fontsize=25)
plt.xlabel('Terror Group Name',fontsize=25)
plt.ylabel('Number of Attacks',fontsize=25)
```

Taliban is the most infamous Terrorist Organization according to Number of attacks
